```python
import os
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain.schema import StrOutputParser
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_neo4j import Neo4jGraph
from neo4j import GraphDatabase
```

```python
def load_properties(path):
    data = {}
    with open(path) as f:
        for line in f:
            if "=" in line:
                k, v = line.split("=", 1)
                data[k.strip()] = v.strip()
    return data
```

```python
props = load_properties("openai_key.txt")

os.environ["NEO4J_URI"] = props["NEO4J_URI"]
os.environ["OPENAI_KEY"] = props["OPENAI_KEY"]
os.environ["NEO4J_USERNAME"] = props["NEO4J_USERNAME"]
os.environ["NEO4J_PASSWORD"] = props["NEO4J_PASSWORD"]
#os.environ["NEO4J_DATABASE"] = props["NEO4J_DATABASE"]
```

```python
chat_llm = ChatOpenAI(
    openai_api_key=os.getenv("OPENAI_KEY")
)
```

```python
from langchain_neo4j import Neo4jChatMessageHistory


def get_memory(session_id):
    return Neo4jChatMessageHistory(session_id=session_id, graph=graph)
```

```python
prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are a surfer dude, having a conversation about the surf conditions on the beach. Respond using surfer slang.",
        ),
        ("system", "{context}"),
        MessagesPlaceholder(variable_name="history"),
        ("human", "{question}"),
    ]
)
```

```python
chat_chain = prompt | chat_llm | StrOutputParser()
```

```python
print(type(chat_chain))
```

```python
chat_with_message_history = RunnableWithMessageHistory(
    runnable=chat_chain,
    get_session_history=get_memory,
    input_messages_key="question",
    history_messages_key="history",
    return_messages=True
)
```

```python
current_weather = """
    {
        "surf": [
            {"beach": "Fistral", "conditions": "6ft waves and offshore winds"},
            {"beach": "Bells", "conditions": "Flat and calm"},
            {"beach": "Watergate Bay", "conditions": "3ft waves and onshore winds"}
            {"beach": "Brighton", "conditions": "4ft waves and Calm"}
        ]
    }"""
```

```python
while (question := input("> ")) != "exit":

    response = chat_with_message_history.invoke(
        {
            "context": current_weather,
            "question": question,

        },
        config={
            "configurable": {"session_id": SESSION_ID}
        }
    )

    print(response)
```
